```python
###### function to create the test error for model without basis expansion
# *********************************Parameters***********************************
# S: subset
# x_test: new feature 
# t_train: t-values from the training set
# t_test: t-values from the test set
# M: test length
# length: capcity

def test_err(S, x_test, t_train, t_test, M, length):
    print("Error S(", length+1, ")")
    
    one_col = np.ones(N)
    t_col = np.array([t_train])
    
    # assigning X matrix
    X = np.insert(S[:N, 1:2], 0, one_col, axis = 1)

    for i in range(1, length+1, 1):
        X = np.insert(S[:N, i+1:i+2], 0, X.T, axis = 1)
    
    #*******Debugging print statementes**********
    #print(X)
    #print(X.shape)
    
    #*******Definition comments**********
    # W = (X^T*X)^-1*(X^T*t)
    # W_partA = (X^T*X)^-1
    # W_partB = (X^T*t)
    
    W_partA = np.dot(X.T, X)
    #*******Debugging print statementes**********
    #print("Before Inv:", W_partA)
    #print("Before Inv:", W_partA.shape)
    
    W_partA = np.linalg.inv(W_partA)
    #*******Debugging print statementes**********
    #print("After Inv:", W_partA)
    #print("After Inv:", W_partA.shape)
    
    W_partB = np.dot(X.T, t_col.T)
    #*******Debugging print statementes**********
    #print(W_partB)
    #print(W_partB.shape)
    
    W = np.dot(W_partA, W_partB)
    #*******Debugging print statementes**********
    #print(W)
    #print(W.shape)
    
    # creating the X and t matrix to find the test error 
    x_test_col = np.ones(M)
    x_test_col = np.insert(x_test[:N, 0:1], 0, x_test_col, axis = 1)
    
    for i in range(1, length+1, 1):
        x_test_col = np.insert(x_test[:N, i:i+1], 0, x_test_col.T, axis = 1)
    
    t_test_col = np.array([t_test])
    
    #*******Definition comments**********
    #print(x_test_col)
    #print(x_test_col.shape)
    #print(t_test_col)
    #print(t_test_col.shape)
    
    #*******Definition comments**********
    # err = (1/N)*(Xw - t)^T*(Xw - t)
    # err_partA = (Xw - t)^T
    # err_partB = (Xw - t)
    
    err_partA = np.dot(x_test_col, W) - t_test_col.T
    #*******Debugging print statementes**********
    #print(err_partA)
    #print(err_partA.shape)
    
    err_partB = np.dot(x_test_col, W) - t_test_col.T
    #*******Debugging print statementes**********
    #print(err_partB)
    #print(err_partB.shape)
    
    err = np.dot(err_partA.T, err_partB)/N
    err = np.sqrt(err)
    #err = 0
    
    return err
```

```python
###### function to train the model using the subset for the model without basis expansion
# *********************************Parameters***********************************
# S: subset
# x_test: new feature 
# t_train: t-values from the training set
# t_test: t-values from the test set
# N: training set length
# M: test set length
# k: capcity

def model_train(S, x_train, x_test, t_train, t_test, N, M, k):
    #print("k = ", k);
    
    one_col = np.array([np.ones(N)])
    t_col = np.array([t_train])
    X = one_col
    #*******Debugging print statementes**********
    #print(one_col)
    #print(one_col.T.shape)
    #print(t_col)
    #print(t_col.shape)
    #print("X shape:",X.shape)
    
    S_len = len(S)
    #print(S[:S_len-M].shape)
    #print(S[:S_len - M, 1:2])
    
    # assigning X matrix
    if(k == 1):
        X = np.insert(x_train, 0, one_col, axis = 1)
    elif(k % 2 == 0): # for the even k-values greater than 1
        for i in range (1, k, 1):
            X = np.insert(S[:S_len - M, i:i+1].T, 0, X, axis = 0)
        X = np.insert(x_train**k, 0, X, axis = 1)
    elif(k % 2 != 0): # for the oddd k-values greater than 1
        for i in range (1, k, 1):
            X = np.insert(S[:S_len - M, i:i+1].T, 0, X, axis = 0)
        X = np.insert(x_train, 0, X, axis = 1)

    #*******Debugging print statementes**********
    #print(X)
    #print(X.shape)
    
    #*******Definition comments**********
    # W = (X^T*X)^-1*(X^T*t)
    # W_partA = (X^T*X)^-1
    # W_partB = (X^T*t)
    
    W_partA = np.dot(X.T, X)
    #*******Debugging print statementes**********
    #print("Before Inv:", W_partA)
    #print("Before Inv:", W_partA.shape)
    
    W_partA = np.linalg.inv(W_partA)
    #*******Debugging print statementes**********
    #print("After Inv:", W_partA)
    #print("After Inv:", W_partA.shape)
    
    W_partB = np.dot(X.T, t_col.T)
    #*******Debugging print statementes**********
    #print(W_partB)
    #print(W_partB.shape)
    
    W = np.dot(W_partA, W_partB)
    #*******Debugging print statementes**********
    #print(W)
    #print(W.shape)
    
    x_test_col = np.ones(M)
    #*******Debugging print statementes**********
    #print(x_test_col)
    #print(x_test_col.shape)

    # creating the X and t matrix to find the test error 
    if(k == 1):
        x_test_col = np.insert(x_test, 0, x_test_col, axis = 1)
    elif(k >= 2): # for k-values greater than 2
        for i in range (1, k, 1):
            x_test_col = np.insert(S[:M - S_len, i: i+1].T, 0, x_test_col.T, axis = 0)
            x_test_col = x_test_col.T
        x_test_col = np.insert(x_test, 0, x_test_col.T, axis = 1)
    
    t_test_col = np.array([t_test])
    #*******Debugging print statementes**********
    #print(x_test_col)
    #print(x_test_col.shape)
    #print(t_test)
    #print(t_test.shape)
    
    #*******Definition comments**********
    # err = (1/N)*(Xw - t)^T*(Xw - t)
    # err_partA = (Xw - t)^T
    # err_partB = (Xw - t)
    
    err_partA = np.dot(x_test_col, W) - t_test_col.T
    #*******Debugging print statementes**********
    #print(err_partA)
    #print(err_partA.shape)
    
    err_partB = np.dot(x_test_col, W) - t_test_col.T
    #*******Debugging print statementes**********
    #print(err_partB)
    #print(err_partB.shape)
    
    err = np.dot(err_partA.T, err_partB)/N
    err = np.sqrt(err)
    #err = 0

    #print(X)
    #print(X.shape)
    return err
```

```python
###### function to train the model using the subset for the model with basis expansion x^n
# *********************************Parameters***********************************
# S: subset
# x_test: new feature 
# t_train: t-values from the training set
# t_test: t-values from the test set
# N: training set length
# M: test set length
# k: capcity

def model_train_bases1(S, x_train, x_test, t_train, t_test, N, M, k):
    #print("k = ", k);
    
    one_col = np.array([np.ones(N)])
    t_col = np.array([t_train])
    X = one_col
    #*******Debugging print statementes**********
    #print(one_col)
    #print(one_col.T.shape)
    #print(t_col)
    #print(t_col.shape)
    #print("X shape:",X.shape)
    
    S_len = len(S)
    #print(S[:S_len-M].shape)
    #print(S[:S_len - M, 1:2])
    
    # assigning X matrix
    if(k == 1):
        X = np.insert(x_train, 0, one_col, axis = 1)
    elif(k % 2 == 0): # for the even k-values greater than 1
        for i in range (1, k, 1):
            X = np.insert(S[:S_len - M, i:i+1].T**i, 0, X, axis = 0)
        X = np.insert(x_train**k, 0, X, axis = 1)
    elif(k % 2 != 0): # for the odd k-values greater than 1
        for i in range (1, k, 1):
            X = np.insert(S[:S_len - M, i:i+1].T**i, 0, X, axis = 0)
        X = np.insert(x_train**k, 0, X, axis = 1)

    #*******Debugging print statementes**********
    #print(X)
    #print(X.shape)
    
    
    
    #*******Definition comments**********
    # W = (X^T*X)^-1*(X^T*t)
    # W_partA = (X^T*X)^-1
    # W_partB = (X^T*t)
    
    W_partA = np.dot(X.T, X)
    #*******Debugging print statementes**********
    #print("Before Inv:", W_partA)
    #print("Before Inv:", W_partA.shape)
    
    W_partA = np.linalg.inv(W_partA)
    #*******Debugging print statementes**********
    #print("After Inv:", W_partA)
    #print("After Inv:", W_partA.shape)
    
    W_partB = np.dot(X.T, t_col.T)
    #*******Debugging print statementes**********
    #print(W_partB)
    #print(W_partB.shape)
    
    W = np.dot(W_partA, W_partB)
    #*******Debugging print statementes**********
    #print(W)
    #print(W.shape)
    
    x_test_col = np.ones(M)
    #*******Debugging print statementes**********
    #print(x_test_col)
    #print(x_test_col.shape)
    
    # creating the X and t matrix to find the test error 
    if(k == 1):
        x_test_col = np.insert(x_test, 0, x_test_col, axis = 1)
    elif(k >= 2): # for k-values greater than 2
        for i in range (1, k, 1):
            x_test_col = np.insert(S[:M - S_len, i: i+1].T**i, 0, x_test_col.T, axis = 0)
            x_test_col = x_test_col.T
        x_test_col = np.insert(x_test**k, 0, x_test_col.T, axis = 1)
    
    t_test_col = np.array([t_test])
    #*******Debugging print statementes**********
    #print(x_test_col)
    #print(x_test_col.shape)
    #print(t_test)
    #print(t_test.shape)
    
    #*******Definition comments**********
    # err = (1/N)*(Xw - t)^T*(Xw - t)
    # err_partA = (Xw - t)^T
    # err_partB = (Xw - t)
    
    err_partA = np.dot(x_test_col, W) - t_test_col.T
    #*******Debugging print statementes**********
    #print(err_partA)
    #print(err_partA.shape)
    
    err_partB = np.dot(x_test_col, W) - t_test_col.T
    #*******Debugging print statementes**********
    #print(err_partB)
    #print(err_partB.shape)
    
    err = np.dot(err_partA.T, err_partB)/N
    err = np.sqrt(err)
    #err = 0

    #print(X)
    #print(X.shape)
    return err
```
